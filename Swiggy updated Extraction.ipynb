```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

```python
links=[]
names=[]

def getlink(pgno):
    url=f"https://www.swiggy.com/coimbatore?page={pgno}?"
    r=requests.get(url, headers=headers)
    soup=BeautifulSoup(r.text,'html.parser')
    body=soup.find_all("div",{"class": "_3XX_A"})
    for item in body:
        link=item.find('a',{'class':'_1j_Yo'})
        links.append(link.get('href'))
        print(link.get('href'))
        name=item.find('div',{'class':'nA6kb'})
        names.append(name.text)
```

```python
for i in range(1,18,1):
    print(i)
    getlink(i)
```

```python
#complete link and append to csv
for i in range(0,len(links),1):
    links[i]=("https://www.swiggy.com"+links[i])
df={"Hotel":names,
    "Link":links}
df=pd.DataFrame(df)
df.drop_duplicates("Link",keep="first",inplace=False)
print(df)
```
